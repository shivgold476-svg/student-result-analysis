## STUDENT RESULT ANALYSIS

```python
import pandas as pd
```

```python
marks = pd.DataFrame({
    "StudentID":[101,102,103,104,105],
    "Maths":[98,89,79,80,92],
    "Science":[79,82,84,91,80],
    "Computer":[69,99,81,79,90],
    "English":[79,69,82,93,80]
})
```

```python
students=pd.DataFrame({
    "StudentID":[101,102,103,104,105],
    "Name":["Ram","Shyam","Radha","Rudra","Shivangi"],
    "Age":[18,19,21,18,20]

})
```

```python
data = pd.merge(students,marks)
print(data)
```

```python
data["Total"]=data["Maths"]+data["Science"]+data["Computer"]+data["English"]
print(data)
```

```python
data["Percentage"] = data["Total"]/4
print(data)
```

```python
def get_grade(percentage):
    if percentage >= 90:
        return "A"
    elif percentage >= 80:
        return "B"
    elif percentage >= 70:
        return "C"
    else:
        return "D"
data["Grade "] = data["Percentage"].apply(get_grade)
print(data)
```

```python
average = data["Percentage"].mean()
new_index = len(data)
data.loc[new_index,"Name"] = "Average"
data.loc[new_index,"Percentage"]=average
print(data)
```

```python
topper = data.loc[data["Percentage"].idxmax()]
print("Topper")
print(topper)
```

```python
data.to_csv("output.csv",index = False)
print(data)
```
